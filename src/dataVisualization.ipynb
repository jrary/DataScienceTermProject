```python
#이상치 찾기 - Turkey Fences
import numpy as np
import matplotlib.pyplot
df = pd.read_csv('./assets/output/visitors.csv')
df['visitor'] = df['visitor'].str.replace(',', '').astype(float)
df = df[df['visitor'] <= 20000]

def find_outlier_Turkey(data):
    q1, q3 = np.percentile(data,[25,75])

    iqr = q3 - q1
    
    lower_bound = q1 - (iqr*1.5)
    upper_bound = q3 + (iqr*1.5)

    mask = np.where((data>upper_bound)|(data<lower_bound))

    return mask

outliers =find_outlier_Turkey(df['visitor'])[0]
print(len(outliers))
outliers

```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('./assets/output/finalDataset.csv')
df.dropna()
def findOutlier(feat):
    
    fig = plt.figure()  # 그래프 크기 설정
    subPlot = fig.add_subplot(1,2,1)
    subPlot.hist(df[feat].values, bins=10)  # 히스토그램 그리기
    subPlot.set_xlabel("Value")  # x축 레이블 설정
    subPlot.set_ylabel("Frequency")  # y축 레이블 설정
    fig.suptitle(f"Histogram of {feat}")  # 그래프 제목 설정
    subPlot2 = fig.add_subplot(1,2,2)
    subPlot2.boxplot(df[feat].values)  # 박스 플롯 그리기
    plt.show()  # 그래프 출력)

for col in df.columns:
    if(col == 'date'):
        continue
    findOutlier(col)
```
