```python

import numpy as np
df = pd.read_csv('./assets/output/visitors.csv')
df['visitor'] = df['visitor'].str.replace(',', '').astype(float)
fil_df = df[df['visitor'] <= 20000]

def find_outlier_z(data):
    threshold = 3

    mean = np.mean(data)
    std = np.std(data)

    z_score = [(y-mean)/std for y in data]

    masks = np.where(np.abs(z_score)>threshold)

    return masks

outliers = find_outlier_z(fil_df['visitor'])
outliers
```
