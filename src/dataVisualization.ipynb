```python
df = pd.read_csv('./assets/output/visitors.csv')
df['visitor'] = df['visitor'].str.replace(',', '').astype(float)
# 3 
def find_outlizer_z(data):
    threshold = 3
    
    mean = np.mean(data)
    std = np.std(data)
    
    z_score = [(y - mean)/std for y in data]
    
    masks = np.where(np.abs(z_score)>threshold)
    
    return masks

outliers = find_outlizer_z(df['visitor'])[0]
```

```python
outliers
```

```python
idxs = df['visitor'].index.values

result = []

for idx in idxs:
    if idx in outliers :
        result.append(True)
    else :
        result.append(False)
#outlier        
df['visitor'][result].describe()
```
